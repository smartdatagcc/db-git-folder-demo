
# Databricks Magic Commands

- %python, %sql, %scala, %r: switch to a different languages for a specific cell
- %md: Markdown for documenting notebooks
- %fs: Run File system commands
- %sh: Run shell commands (Driver Node only)
- %pip: Install Python libraries
- %run: Include/Import another notebook into the current notebook


### %fs: Run file system commands

```python
# To list out all the folders in the root
%fs ls /

# To list out all the files in the root
%fs ls /databricks-datasets/

# To copy a file use 'cp' command and to move a file from one folder to another folder use 'mv' command
```

### %sh: Run shell command (Driver node only)

```python
# To list out all the processes running
%sh ps
```

### %pip: Install Python libraries

```python
# To list out all python libraries
%pip list
```
