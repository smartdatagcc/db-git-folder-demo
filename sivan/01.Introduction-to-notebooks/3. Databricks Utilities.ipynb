```python
items = dbutils.fs.ls('dbfs:/databricks-datasets/')
```

```python
foldersCount = len([item for item in items if item.name.endswith('/')])
print(foldersCount)
```

```python
fileCount = len([item for item in items if not item.name.endswith('/')])
print(fileCount)
```
