### 4. Select File Meta Data

```sql
select input_file_name() as file_path, -- Depricated from databricks runtime 13.3 LTS onwards
--_metadata,
_metadata.file_path as file_path_new,
*
from json.`/Volumes/gizmobox_sivan/landing/operational_data_volume/customers`
```

### 5. Register Files in Unity Catalog using Views

```sql
CREATE OR REPLACE VIEW gizmobox_sivan.bronze.v_Customers
AS
SELECT *,
_metadata.file_path as file_path
from json.`/Volumes/gizmobox_sivan/landing/operational_data_volume/customers`
```

```sql
select * from gizmobox_sivan.bronze.v_customers;
```

### 6. Create Temporary View

```sql
CREATE OR REPLACE TEMPORARY VIEW tv_Customers
AS
SELECT *,
_metadata.file_path as file_path
from json.`/Volumes/gizmobox_sivan/landing/operational_data_volume/customers`
```

```sql
select * from tv_Customers
```

###7. Create Global Temporary View

```sql
CREATE OR REPLACE GLOBAL TEMPORARY VIEW gtv_Customers
AS
SELECT *,
_metadata.file_path as file_path
from json.`/Volumes/gizmobox_sivan/landing/operational_data_volume/customers`
```

```sql
SELECT * FROM global_temp.gtv_Customers
```
