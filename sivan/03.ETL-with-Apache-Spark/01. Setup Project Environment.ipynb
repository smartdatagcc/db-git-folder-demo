### Set-up the project environment for Gizmobox Data lakehouse

### 1. Access the container gizmobox

```sql
%fs ls 'abfs://gizmoboxsivan@ourdbpracticeextdl.dfs.core.windows.net/landing/operational_data/'
```

### 2. Create External Location

```sql
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS databricks_practice_ext_dl_gizmoboxsivan
    URL 'abfss://gizmoboxsivan@ourdbpracticeextdl.dfs.core.windows.net/'
    WITH (STORAGE CREDENTIAL `databricks-practice-ext-sc`)
    COMMENT 'External Location for the Gizmobox Datalake house'
```

### 3. Create the catalog - gizmobox-Sivan

```sql
CREATE CATALOG IF NOT EXISTS gizmobox_sivan
    MANAGED LOCATION 'abfss://gizmoboxsivan@ourdbpracticeextdl.dfs.core.windows.net/gizmoboxsivan'
    COMMENT 'This is the catalog for the gizmobox-sivan data lake house';
```

### See the existing catalogs

```sql
SHOW CATALOGS
```

### 4. Create Schemas

```sql
%sql
-- To See the Current Used Catalog
SELECT current_catalog()
```

```sql
-- To switch the current catalog to a specific Catalog
USE CATALOG gizmobox_sivan
```

```sql
-- To Create a Schema
CREATE SCHEMA IF NOT EXISTS landing
    COMMENT 'This is the landing zone for the gizmobox-sivan data lake house'
    MANAGED LOCATION 'abfss://gizmoboxsivan@ourdbpracticeextdl.dfs.core.windows.net/gizmoboxsivan/landing'
```

```sql
-- To See the existing schemas in the current catalog
SHOW SCHEMAS
```

```sql
CREATE SCHEMA IF NOT EXISTS bronze
    COMMENT 'This is the bronze folder in the gizmobox-sivan data lake house'
    MANAGED LOCATION 'abfss://gizmoboxsivan@ourdbpracticeextdl.dfs.core.windows.net/gizmoboxsivan/bronze';

CREATE SCHEMA IF NOT EXISTS silver
    COMMENT 'This is the silver folder in the gizmobox-sivan data lake house'
    MANAGED LOCATION 'abfss://gizmoboxsivan@ourdbpracticeextdl.dfs.core.windows.net/gizmoboxsivan/silver';

CREATE SCHEMA IF NOT EXISTS gold
    COMMENT 'This is the gold folder in the gizmobox-sivan data lake house'
    MANAGED LOCATION 'abfss://gizmoboxsivan@ourdbpracticeextdl.dfs.core.windows.net/gizmoboxsivan/gold';
```

### 5. Create Volume

```sql
-- Use Specific Catalog and Schema
USE CATALOG gizmobox_sivan;
USE SCHEMA landing;

CREATE EXTERNAL VOLUME IF NOT EXISTS operational_data_volume
    COMMENT 'This is the operational data volume for the gizmobox-sivan data lake house'
    LOCATION 'abfss://gizmoboxsivan@ourdbpracticeextdl.dfs.core.windows.net/landing/operational_data/';


```
