# Databricks Magic Commands
- **%python, %sql, %scala, %r**: switch to a different languages for a specific cell

- **%md** is used to create markdown cells. It allows you to add documentation

- **%fs** magic command lets you interact with the Databricks file system. This command lets
you execute basic file system commands such as listing files, copying files from one folder
to another, moving a file from one folder to another folder etc.

- **%sh** magic command allows you to run shell commands directly within your notebook. For 
example, you can look to see all the processes that are running on the cluster and install
a new package etc. using this magic command.

- **%pip** magic command is used to manage Python libraries within the notebook environment.
You can install Python packages from PyPi or other sources directly into your Databricks
notebook environment using the %pip magic command.

- **%run** magic command  is used to include or import another notebook into the
current notebook. For example, you can define environment variables and common functions
in their own notebooks and include them in every notebook using the %run command rather
than duplicating the code everywhere. They basically help us modularize the code, avoid
any duplication and keep all the notebooks maintainable.   
This is one of the most commonly used magic commands in the industry.
