
### Databricks Utilities

- **File system utilities :** allows us to access Databricks file system from a notebook, and
you can use various file system level operations.  It's exactly the same as the **%fs** magic command, but the Databricks utilities are more flexible to use than the magic commands

- **Secret utilities :**  allow us to get secret values from secrets stored in SecretScope backed by Databricks or Azure Key Vault.

- **Widget utilities :** allows us to parameterize notebooks so that a calling notebook or another application, for example, Azure Data Factory Pipeline, can pass a parameter value to the notebook at runtime. This is really useful to make notebooks reusable.

- **Notebook workflow utilities** : allow us to invoke one notebook from another and chain them together. This is no longer recommended, and Databricks wants us to avoid this if possible, as Databricks offers another tool called workflows or Databricks workflows, and that's constantly being improved and easier to use, and we can monitor the workflow jobs very easily as well.

#### File system utilities

```python
%fs ls '/'
```

```python
display(dbutils.fs.ls('/'))
```

```python
%fs ls dbfs:/databricks-datasets/
```

```python
display(dbutils.fs.ls('dbfs:/databricks-datasets/'))
```

```python
items= dbutils.fs.ls('dbfs:/databricks-datasets/')
fileCount = len([item for item in items if item.name.endswith('/')])
#print('foldersCount:', foldersCount)
print(f"Folders count: {fileCount}")

fileCount = len([item for item in items if not item.name.endswith('/')])
#print('fileCount:', fileCount)
print(f"Files count: {fileCount}")
```
