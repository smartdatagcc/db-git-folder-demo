##### Create External Location

```python
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS databricks_practice_ext_dl_practice
    URL 'abfss://practice@ourdbpracticeextdl.dfs.core.windows.net/'
    WITH (STORAGE CREDENTIAL `databricks-practice-ext-sc`)
    COMMENT 'External Location for practice purpose'
```
